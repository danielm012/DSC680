```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hotDogWinnerData = pd.read_excel("C:/Users/danie/OneDrive/DSC-640/week1_2/ex1-2/hotdog-contest-winners.xlsm")
hotDogPlaces = pd.read_excel("C:/Users/danie/OneDrive/DSC-640/week1_2/ex1-2/hotdog-places.xlsm")
obamaApprovaData = pd.read_excel("C:/Users/danie/OneDrive/DSC-640/week1_2/ex1-2/obama-approval-ratings.xls")
```

```python
sns.relplot(x="Winner" , y="New record" , kind="line", data=hotDogWinnerData).fig.set_figwidth(45)
```

```python
sns.catplot(x="Winner", y="Dogs eaten", kind="bar", data=hotDogWinnerData).fig.set_figwidth(45)
```

```python
plt.pie(hotDogPlaces.sum(axis=0), labels=hotDogPlaces.columns)
```

```python
#sns.replot(x= , y= , kind="line", data=)
sns.set_color_codes("pastel")

sns.set_color_codes("muted")
plt.subplots(figsize=(40, 5))
sns.barplot(x="Issue", y="Approve", data=obamaApprovaData,
            label="Issue", color="g")
sns.barplot(x="Issue", y="Disapprove", data=obamaApprovaData,
            label="Issue", color="y")
```
