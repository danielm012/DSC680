```python
# giving title to the plot
plt.title("Income Data")
      
# giving X and Y labels
plt.xlabel("index")
plt.ylabel("Income")
    
plt.bar(income_json.index, income_json['Income'])
```

```python
# giving title to the plot
plt.title("Income Data")
      
# giving X and Y labels
plt.xlabel("index")
plt.ylabel("Income")

plt.plot(income_json.index, income_json['Income'])
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct

plt.pie(income_json['# of kids'], labels = income_json['State'], autopct=make_autopct(income_json['# of kids']))
plt.show()
```
