```python
def changeColumnsToCategorical(df, keys, columns):
    for column in columns:
        for index in range(0, len(df)):
            if(df[column][index] in keys):
                df.loc[index, column] = keys.get(df[column][index])
    return df
```

```python
companiesMLDf = changeColumnsToCategorical(companiesMLDf, raceKey, ['race'])
companiesMLDf = changeColumnsToCategorical(companiesMLDf, genderKey, ['gender'])
companiesMLDf = changeColumnsToCategorical(companiesMLDf, jobKey, ['job_category'])
companiesMLDf = changeColumnsToCategorical(companiesMLDf, companyKey, ['company'])
```

```python
companyColumn = ['23andMe', 'Adobe', 'Airbnb', 'Apple', 'Cisco', 'eBay', 'Facebook',
       'Google', 'HP Inc.', 'HPE', 'Intel', 'Intuit', 'LinkedIn', 'Lyft',
       'MobileIron', 'Nvidia', 'Pinterest', 'Salesforce', 'Square',
       'Twitter', 'Uber', 'View']

raceColumn = ['American_Indian_Alaskan_Native', 'Asian',
       'Black_or_African_American', 'Latino',
       'Native_Hawaiian_or_Pacific_Islander',
       'Two_or_more_races', 'White']

genders = {
   'female', 'male'
}
```

```python
def findProbability(df1,  companies, races):
    probArr = []
    indexArr = []
    total = 0
    for company in companies:
        for race in races:
            for gender in genders:
                dfNew = df1.query('company == "%s" and race == "%s" and gender == "%s" and job_category != "Totals" and job_category != "Previous_totals" and race != "Overall_totals"' %(company, race, gender))
                dfNew['count'] = dfNew['count'].astype('int32')
                for index in range(0, len(dfNew)):
                    idx = int(dfNew.index[index])
                    divisor = int(df1.query('company == "%s" and race == "Overall_totals" and job_category != "Totals" and job_category != "Previous_totals"' %(company)).reset_index(drop=True)['count'][index%10])
                
                    print("index: " + str(dfNew.index[index]) + ' company == "%s" and race == "%s"' %(company, race) + " Job: " + dfNew['job_category'][idx] + " \n\t\tcurrent: " + str(dfNew['count'][idx]) + " divisor: " + str(divisor))
                   
                    indexArr.append(dfNew.index[index])
                    
                    if(dfNew['count'][idx] == 0 or divisor == 0):
                        probArr.append(0)
                    else:
                        probArr.append( (dfNew['count'][idx] / divisor) * 1)
                total += len(dfNew)
    return pd.DataFrame(probArr, indexArr)
```

```python
targetColumn = ['company', 'race', 'gender', 'job_category']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(companiesMLDf[targetColumn], companiesMLDf['probability'], train_size=0.80)
```

```python
# Multiple classifier trial and error to find the best fit
from sklearn.metrics import accuracy_score, log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import *
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import *
from sklearn.svm import *
from sklearn.ensemble import *

classifiers = [
    ExtraTreesClassifier(),
    RidgeClassifier(),
    RandomForestClassifier(),
    BaggingClassifier(),
    KNeighborsClassifier()
]

# Logging for Visual Comparison
log_cols=["Classifier", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)

for clf in classifiers:
    clf.fit(x_train, y_train)
    name = clf.__class__.__name__
    
    print("="*30)
    print(name)
    
    print('****Results****')
    train_predictions = clf.predict(x_test)
    acc = accuracy_score(y_test, train_predictions)
    print("Accuracy: {:.4%}".format(acc))
    
print("="*30)
```
